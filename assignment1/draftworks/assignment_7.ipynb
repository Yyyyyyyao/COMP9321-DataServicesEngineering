```python
def find_duplicates(x):
    # find duplicate countries
    temp1 = x[x.duplicated(keep=False)]
    # uniq the duplicate countries
    temp3 = temp1.drop_duplicates()
    # put them into a comma seperate list
    l = temp3.to_list()
    return l
```
