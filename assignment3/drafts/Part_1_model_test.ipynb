```python
import numpy as np
import pandas as pd
import sklearn
import datetime
import os
import json
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn import preprocessing
from sklearn.model_selection import train_test_split, KFold
import math
```

```python
raw_training = pd.read_csv('training_set.csv')
raw_validation = pd.read_csv('validation_set.csv')
```

```python
raw_training['budget'] = raw_training['budget'].apply(lambda x: math.log(x))
```

```python
raw_training
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64'] #so that easy for us to perform  train and test
df_train=raw_training.select_dtypes(include=numerics)
df_validate = raw_validation.select_dtypes(include=numerics)
```

```python
df_train
```

```python
train_y = df_train["revenue"].apply(lambda x: math.log(x))
train_X = df_train.drop(axis=1, columns=["movie_id", "rating", "revenue"])

# train_X = df_train[['runtime', 'budget', 'release_date_year', 'release_date_quarter']]
# train_X = df_train[["budget", "runtime"]]
```

```python
train_X
```

```python
validate_y = df_validate["revenue"]
validate_X = df_validate.drop(axis=1, columns=["movie_id", "rating", "revenue"])
# validate_X = df_validate[["budget", "runtime"]]
```

```python
normalized_X_train = preprocessing.normalize(train_X)
# normalized_y_train = preprocessing.normalize(train_y)

normalized_X_val = preprocessing.normalize(validate_X)
# normalized_y_val = preprocessing.normalize(validate_y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_X, train_y, random_state = 1)
#linear regression
lm = LinearRegression() #our 6th model
lm.fit(X_train, y_train)
lm_preds = lm.predict(X_test)
a = mean_squared_error(lm_preds, y_test)
print ("R-Squared is:", r2_score(lm_preds, y_test))


# from sklearn.svm import SVR

# svm_model = SVR()
# svm_model.set_params(**{'kernel': 'poly', 'degree': 1, 'C': 5000,
#                         'gamma': 'scale', 'coef0': 0.0, 'tol': 0.001, 'epsilon': 10})
# # svr_poly = SVR(kernel='poly', C=100, gamma='auto', degree=3, epsilon=.1,coef0=1)
# svm_model.fit(X_train, y_train)
# pred = svm_model.predict(X_test)
# a = mean_squared_error(pred, y_test)
```

```python
res = y_test.apply(lambda x: math.exp(x))
pres = np.exp(lm_preds)
```

```python
import matplotlib.pyplot as plt
pred = np.array(lm_preds)
GT = np.array(y_test)
fig, ax = plt.subplots(figsize=(15,20))
for i in range(len(pred)):
    ax.plot([i,i],[pred[i], GT[i]], c="k", linewidth=0.5)
ax.plot(pred,'o', label='Prediction', color = 'g')
ax.plot(GT,'^', label='Ground Truth', color = 'r' )

ax.set_xlim((-1,20))
plt.xlabel('a')
plt.ylabel('b')
plt.title('test')

plt.legend()             
plt.show()
```

```python
print ("R-Squared is:", r2_score(pres, res))
```
